# Part 1: Bond Pricing & Yield Curve Bootstrapping

## 1.1 Bond Pricing

A fixed-coupon bond pays periodic coupons $C$ and a face value $F$ at maturity.
Given a yield-to-maturity $y$ and $m$ coupon payments per year over $n$ years,
the price is the present value of all cash flows:

$$P = \sum_{t=1}^{n \cdot m} \frac{C/m}{(1 + y/m)^t} + \frac{F}{(1 + y/m)^{n \cdot m}}$$

## 1.2 Sensitivity Measures

**Macaulay Duration** — weighted average time of cash flows:
$$D_{Mac} = \frac{1}{P} \sum_{t=1}^{T} t \cdot \frac{CF_t}{(1 + y/m)^t}$$

**Modified Duration** — price sensitivity to yield:
$$D_{mod} = \frac{D_{Mac}}{1 + y/m} \implies \frac{\Delta P}{P} \approx -D_{mod} \cdot \Delta y$$

**DV01** — dollar change per 1 basis point move:
$$DV01 = D_{mod} \cdot P \cdot 0.0001$$

**Convexity** — second-order correction:
$$C = \frac{1}{P} \cdot \frac{\partial^2 P}{\partial y^2}$$

Full price approximation:
$$\Delta P \approx -D_{mod} \cdot P \cdot \Delta y + \frac{1}{2} \cdot C \cdot P \cdot (\Delta y)^2$$

```python
import numpy as np

def bond_price(face: float, coupon_rate: float, ytm: float, 
               n_years: int, freq: int = 2) -> float:
    """
    Price a fixed-coupon bond.
    
    Parameters
    ----------
    face        : float  — face/par value
    coupon_rate : float  — annual coupon rate (e.g. 0.085 for 8.5%)
    ytm         : float  — yield to maturity, annual (e.g. 0.092)
    n_years     : int    — years to maturity
    freq        : int    — coupon payments per year (default: 2 semiannual)
    
    Returns
    -------
    float — bond price
    """
    n_periodos = n_years * freq

    t = np.arange(1, n_periodos + 1) # opera sobre todo el vector de una vez
    c = np.full(n_periodos, coupon_rate * face / freq) # vector de cupones
    c[-1] += face # agregar el valor nominal al último cupón
    discount_factors = 1 / (1 + ytm / freq) ** t # vector de factores de descuento
    price = np.sum(c * discount_factors) # precio del bono
    
    return price
    
```

```python
bond_price(1000, 0.085, 0.092, 5, 2)
```
